```python
from tkinter import *
from tkinter import messagebox

def newtask():
    task = my_entry.get()
    if task != "":
        lb.insert(END, task)
        my_entry.delete(0, "end")
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

def deletetask():
    lb.delete(ANCHOR)

ws = Tk()
ws.geometry("500x450+500+200")
ws.title("PYTHON GUIDES")
ws.config(bg="#223441")
ws.resizable(width=False, height=False)

frame = Frame(ws)
frame.pack(pady=10)

lb = Listbox(
    frame,
    width=25,
    height=8,
    font=("Times", 18),
    bd=0,
    fg='#464646',
    highlightthickness=0,
    selectbackground='#a6a6a6',
    activestyle="none",
)
lb.pack(side=LEFT, fill=BOTH)

task_list = [
    "eat apple",
    "drink water",
    "go to gym",
]

for item in task_list:
    lb.insert(END, item)

my_entry = Entry(
    ws,
    font=("Helvetica", 24),
    width=18,
    bg="#394867",
    fg="#ffffff",
)
my_entry.pack(pady=20)

button_frame = Frame(ws)
button_frame.pack(pady=20)

add_btn = Button(
    button_frame,
    text="Add Task",
    font=("Helvetica", 14),
    bg="#00ADB5",
    fg="#ffffff",
    padx=20,
    pady=10,
    command=newtask,
)
add_btn.grid(row=0, column=0)

delete_btn = Button(
    button_frame,
    text="Delete Task",
    font=("Helvetica", 14),
    bg="#D90429",
    fg="#ffffff",
    padx=20,
    pady=10,
    command=deletetask,
)
delete_btn.grid(row=0, column=1)

ws.mainloop()
```
